```python
for year in range(2020, 2021):
    for week in range(17, 18):
        r = requests.get('https://www.worldfootball.net/schedule/esp-primera-division-'
                         +str(year)+'-'+str(year+1)+'-spieltag/'+str(week)+'/')
        r1 = requests.get('https://www.worldfootball.net/schedule/esp-primera-division-'
                         +str(year)+'-'+str(year+1)+'-spieltag/'+str(week-1)+'/')

        html = BS(r.content, 'html.parser')
        html1 = BS(r1.content, 'html.parser')

        club1 = []
        club2 = []

        i = 0
        for el in html.select('.box')[0].select('tr'):
            if i == 10:
                break

            i += 1

            tds = el.select('td')
                
            try:

                club1.append(tds[2].select('a')[0].text)
                club2.append(tds[4].select('a')[0].text)
            except:
                continue

        dct = {}

        i = 0
        for el in html1.select('.box')[1].select('tr'):
            i += 1
            if i == 1:
                continue

            ###tabclubs.append([])

            tds = el.select('td')

            this_club = tds[2].select('a')[0].text # TEAM NAME

            if not this_club in ranks[year]: 
                ranks[year][this_club] = {}
                while True:
                    try:
                        ranks[year][this_club]['att'] = int(input(this_club+' '+str(year)+'att:'))
                        break
                    except:
                        print('error, try again')
                        continue

                while True:
                    try:
                        ranks[year][this_club]['mid'] = int(input(this_club+' '+str(year)+'mid:'))
                        break
                    except:
                        print('error, try again')
                        continue
                while True:
                    try:
                        ranks[year][this_club]['def'] = int(input(this_club+' '+str(year)+'def:'))
                        break
                    except:
                        print('error, try again')
                        continue
            
            goals = tds[7].text
            goals = goals.split(':')
            this_matches = int(tds[3].text)

            if this_matches == 0:
                dct[this_club] = 9999
            else:
                dct[this_club] = (int(goals[0]) - int(goals[1])) / this_matches

            if i == 21:
                break


        ldf = []
        ii = 0

        for i in range(len(club1)):
            if dct[club1[i]] == 9999 or dct[club2[i]] == 9999:
                continue

            ldf.append([])


            ldf[ii].append(year)
            ldf[ii].append(week)
            ldf[ii].append(club1[i]) ### CLUB1
            ldf[ii].append(club2[i]) ### CLUB2
              
            ldf[ii].append(ranks[year][club1[i]]['att'])
            ldf[ii].append(ranks[year][club1[i]]['mid'])
            ldf[ii].append(ranks[year][club1[i]]['def'])
            ldf[ii].append(ranks[year][club2[i]]['att'])
            ldf[ii].append(ranks[year][club2[i]]['mid'])
            ldf[ii].append(ranks[year][club2[i]]['def'])

            ldf[ii].append(dct[club1[i]])
            ldf[ii].append(dct[club2[i]])

            ii += 1

        print(club1)
            
        week6 = pd.DataFrame(ldf, columns=columns)
        
        X = X.append(week6, ignore_index=True)
```
